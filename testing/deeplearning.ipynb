```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
data = pd.read_csv("diabetes.csv")
```

```python
data.head()
```

```python
x = data.drop(columns=["Outcome"])
y = data["Outcome"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model= Sequential()
```

```python
model.add(Dense(8,input_dim=8,activation='relu')) #input neuron created
model.add(Dense(8,activation='relu')) #hidden layer
model.add(Dense(1,activation='sigmoid')) #softmax #output layer
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(xtrain,ytrain,epochs=500)
```

```python
acc=model.evaluate(xtest,ytest)
print('Accuracy score for test data:', acc)
```
